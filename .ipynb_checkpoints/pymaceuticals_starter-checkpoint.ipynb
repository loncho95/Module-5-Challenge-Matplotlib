```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
# %matplotlib notebook



# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
complete_ds = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID"])

                                                                       
# Display the data table for preview
complete_ds.head()
```

```python
# Checking the number of mice.
number_mice = complete_ds["Mouse ID"].unique().size
print(number_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicateRows = complete_ds.duplicated(['Mouse ID', 'Timepoint'])
complete_ds[duplicateRows]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
ds_duplicated = complete_ds[complete_ds["Mouse ID"]=="g989"]
ds_duplicated
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
ds_duplicated_indexes = ds_duplicated.index
ds_duplicated_indexes

noduplicates_ds = complete_ds.drop(index=ds_duplicated_indexes)
noduplicates_ds

```

```python
# Checking the number of mice in the clean DataFrame.
number_mice = noduplicates_ds["Mouse ID"].unique().size
print(number_mice)

```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_Tumor_Volume = noduplicates_ds.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
mean_Tumor_Volume

median_Tumor_Volume = noduplicates_ds.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
var_Tumor_Volume = noduplicates_ds.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
std_Tumor_Volume = noduplicates_ds.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
sem_Tumor_Volume = noduplicates_ds.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

summary_tumors = {"Mean Tumor Volume": mean_Tumor_Volume,
                    "Median Tumor Volume":  median_Tumor_Volume,
                    "Tumor Volume Variance	":var_Tumor_Volume,
                    "Tumor Volume Std. Dev.": std_Tumor_Volume,
                    "Tumor Volume Std. Err.": sem_Tumor_Volume}
summary_tumors_df = pd.DataFrame(summary_tumors)
summary_tumors_df 


```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

# summary_tumors_df_2v = noduplicates_ds.groupby("Drug Regimen").("Tumor Volume (mm3)").agg(['mean','median','var','std','sem'])
summary_tumors_df_2v = noduplicates_ds.groupby("Drug Regimen").agg({'Tumor Volume (mm3)': ['mean','median','var','std','sem']})
summary_tumors_df_2v
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_times = noduplicates_ds.groupby(["Mouse ID"]).max()["Timepoint"]
max_times

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
Max_times_complete = pd.merge(max_times, noduplicates_ds, how="left", on=["Mouse ID","Timepoint"])
#Max_times_complete = Max_times_complete.["Drug Regimen" == "Capomulin"]
# Max_times_complete[['Mouse ID','Tumor Volume (mm3)','Drug Regimen']][Max_times_complete['Drug Regimen'].isin(['Propriva',
#                                                                                                              'Ramicane'])]

Max_times_complete[['Mouse ID', 'Tumor Volume (mm3)','Drug Regimen']][(Max_times_complete['Drug Regimen']=='Capomulin')|(Max_times_complete['Drug Regimen']=='Ramicane')|(Max_times_complete['Drug Regimen']=='Infubinol')|(Max_times_complete['Drug Regimen']=='Ceftamin')]
Max_times_complete
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drugs: 
    tumor_data = Max_times_complete[Max_times_complete["Drug Regimen"]==drug]["Tumor Volume (mm3)"]
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    iqr = tumor_data.quantile(.75) - tumor_data.quantile(.25)
    lowerbound = tumor_data.quantile(.25) - (1.5*iqr)
    upperbound = tumor_data.quantile(.75) + (1.5*iqr)
    
    # add subset 
    tumor_list.append(tumor_data)
     
    
    # Determine outliers using upper and lower bounds
    print(f"{drug} potential outliers: {tumor_data[(tumor_data < lowerbound) | (tumor_data > upperbound)]}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.figure(figsize=[16,6])
plt.xticks(fontsize=18)
plt.boxplot(tumor_list, flierprops={"markerfacecolor":"red","markersize":30})
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

Tummor_volume_l509 = noduplicates_ds[['Tumor Volume (mm3)']][(noduplicates_ds['Mouse ID']=='l509')]
Timepoint_l509 = noduplicates_ds[['Timepoint']][(noduplicates_ds['Mouse ID']=='l509')]
plt.plot(Timepoint_l509, Tummor_volume_l509)
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of mouse l509")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
noduplicates_ds_capolin = noduplicates_ds[(noduplicates_ds["Drug Regimen"] == "Capomulin")]
noduplicates_ds_capolin_tummor = noduplicates_ds_capolin.groupby(["Mouse ID"]).mean()
x_values = noduplicates_ds_capolin_tummor["Tumor Volume (mm3)"]
y_values = noduplicates_ds_capolin_tummor["Weight (g)"]
plt.scatter(x_values,y_values)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(x_values,y_values)
print(f"The correlation between Tumor Volume (mm3) and Weight (g) is {round(correlation[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
